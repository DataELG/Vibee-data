# Concaténation des bases de données des évenements

```python
import pandas as pd
from ast import literal_eval
import uuid
```

# Import df

```python
df_nantes = pd.read_csv('/OUTPUT/nantes_event.csv')
```

```python
df_nantes.head(2)
```

```python
df_BigCityLife = pd.read_csv('/OUTPUT/output2.csv',sep = ';')
```

```python
df_BigCityLife.head(2)
```

```python
df_BigCityLife.drop(columns='Description',inplace=True)
```

```python
df_BigCityLife.rename(columns={'Résumé' : 'Description'},inplace=True)
```

```python
df_BigCityLife.dropna(subset='Lieu',inplace=True)
```

```python
df_BigCityLife.head()
```

```python
df_BigCityLife['Tags'] = df_BigCityLife['Tags'].apply(lambda x: literal_eval(x) if pd.notnull(x) else [])
df_BigCityLife['Tags'] = df_BigCityLife['Tags'].apply(lambda x: ', '.join(x) if isinstance(x, list) else '')
```

```python
df_BigCityLife.tail()
```

# Concat

```python
df_final = pd.concat([df_BigCityLife, df_nantes])
```

```python
df_final.head()
```

```python
df_final.dropna(how = 'all', inplace = True)
```

```python
df_final['Nom'] = df_final['Nom'].str.strip()
```

```python
df_final['Tags'] = df_final['Tags'].astype('str')
```

```python
df_final['id'] = [str(uuid.uuid4()) for _ in range(len(df_final))]
df_final['IdCreateur'] = [str(uuid.uuid4()) for _ in range(len(df_final))]
```

```python
df_final.info()
```

```python
df_final.head()
```

```python
df_final.info()
```

```python
df_final['CreatedAt'] = pd.to_datetime(df_final['CreatedAt'])
df_final['UpdatedAt'] = pd.to_datetime(df_final['UpdatedAt'])
df_final['DateDebut'] = pd.to_datetime(df_final['DateDebut'])
df_final['DateFin'] = pd.to_datetime(df_final['DateFin'])
```

```python
df_final.info()
```

```python
df_final.rename(columns={
    'CreatedAt' : 'createdAt',
    'UpdatedAt' : 'updatedAt'
},inplace=True)
```

```python
column_list=['id','IdCreateur','DateDebut','DateFin','Lieu','Adresse','Tags','Tarif','Description','Image','Nom','createdAt','updatedAt']
```

```python
df_final = df_final[column_list]
```
