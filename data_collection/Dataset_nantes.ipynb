Import et traitement du dataset des evenements de Nantes

```python
import pandas as pd
from utils.cleaning_functions import create_date_debut, create_date_fin
from datetime import datetime
```

```python
csv_link = '/Users/manu/Desktop/SUP/Semaine Campus/INPUT/244400404_agenda-evenements-nantes-nantes-metropole.csv'
```

```python
df = pd.read_csv(csv_link, sep = ';')
```

# Exploration

```python
df.info()
```

```python
df['Date'].nunique()
```

```python
df.columns
```

```python
# Suppression des espaces insécables
df.columns = df.columns.str.replace('\xa0', ' ', regex=True).str.strip()
```

# Cleaning

```python
df['Lieu : code postal'] = df['Lieu : code postal'].astype('str')
```

```python
df.info()
```

```python
df['Lieu : code postal'] = df['Lieu : code postal'].str.replace('.0', '')
```

```python
colums_to_keep = ['Identifiant', 'Date', 'Date','Heure de début','Nom du lieu','Lieu : adresse', 'Lieu : ville', 'Lieu : code postal',"Nom de l'événement", 'Lien Internet', 'Thème','Catégorie 1',	'Catégorie 2',	'Catégorie 3',	'Catégorie 4','Gratuit','Description', 'Image', 'Précisions sur les tarifs']
```

```python
df = df[colums_to_keep]
```

```python
df.head()
```

```python
df['adresse'] = df['Lieu : adresse'] + ' ' + df['Lieu : code postal'] + ' ' + df['Lieu : ville']
```

```python
df.drop(columns=['Lieu : adresse', 'Lieu : ville', 'Lieu : code postal'], inplace = True)
```

```python
df.head()
```

```python
df['price'] = df['Gratuit'].apply(lambda x : 'GRATUIT' if x == 'oui' else 'PAYANT')
```

```python
df.head(2)
```

```python
df.drop(columns=['Gratuit', 'Précisions sur les tarifs'], inplace=True)
```

```python
df.head(1)
```

```python
cols_to_concat = ['Thème', 'Catégorie 1', 'Catégorie 2', 'Catégorie 3', 'Catégorie 4']

df['tags'] = df[cols_to_concat].apply(
    lambda row: ', '.join(row.dropna().astype(str).map(str.strip)),
    axis=1
)
```

```python
df.head(1)
```

```python
df.drop(columns=cols_to_concat, inplace = True)
```

```python
df.head(2)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dropna(how = 'all',inplace=True)
```

```python
df.info()
```

```python
df.columns.values[1] = 'DateDebut'  
df.columns.values[2] = 'DateFin'   
```

```python
df.rename(columns = {
    'Identifiant' : 'id',
    'Nom du lieu' : 'Lieu',
    'adresse' : 'Adresse',
    'tags' : 'Tags',
    'price' : 'Tarif',
    "Nom de l'événement" : 'Nom'
},inplace=True)
```

```python
df.head(1)
```

```python
# Nettoyage des colonnes (supprimer espaces et standardiser les NaN)
df['Heure de début'] = df['Heure de début'].astype(str).str.strip().replace('nan', None)
```

```python
df['DateDebut'] = df.apply(create_date_debut, axis=1)
df['DateFin'] = df.apply(create_date_fin, axis=1)
```

```python
df.info()
```

```python
df.head(5)
```

```python
df['CreatedAt'] = datetime.now()
df['UpdatedAt'] = datetime.now()
```

```python
df.head()
```

```python
df['id'] = ''
df['IdCreateur'] = ''
```

```python
df.head()
```

```python
df.drop(columns=['Lien Internet','Heure de début'], inplace = True)
```
