# Import librairies

```python
import pandas as pd
from faker import Faker
import random
import uuid
from datetime import datetime, timedelta
```

# Generate fake accounts

```python
# Initialisation de Faker
faker = Faker('fr_FR') 

# Nombre de users à générer
n = 150  

# Fonction pour générer une ligne utilisateur
def generate_fake_user(user_id):
    return {
        'id' : uuid.uuid4(),
        'email': faker.email(),
        'password' : faker.password(),
        'createdAt' : datetime.now(),
        'updatedAt' : datetime.now(),
        'user_id' : user_id,
    }

# Génération des données
data = [generate_fake_user(i) for i in range(1, n + 1)]

# Création du DataFrame
df_account = pd.DataFrame(data)

df_account.drop(columns='user_id', inplace= True)
```

# Generate fake reservations

```python
df_users = pd.read_csv('/Users/manu/Desktop/SUP/Semaine Campus/OUTPUT/bdd_users.csv')
df_events = pd.read_csv('/Users/manu/Desktop/SUP/Semaine Campus/OUTPUT/bdd_final_event_.csv')
```

```python

# S'assurer que DateFin est bien en datetime
df_events['DateFin'] = pd.to_datetime(df_events['DateFin'], errors='coerce')

# Filtrer sur événements passés
today = pd.Timestamp(datetime.today().date())
df_events_passes = df_events[df_events['DateFin'] < today].reset_index(drop=True)


# Génération réservations par user
def random_date(start, end):
    """Génère une date aléatoire entre start et end"""
    return start + timedelta(seconds=random.randint(0, int((end - start).total_seconds())))

# Plage pour createdAt/updatedAt : entre 1 an avant aujourd'hui et aujourd'hui
start_date = today - timedelta(days=365)  # il y a 1 an
end_date = today

reservations = []

for _, user_row in df_users.iterrows():
    num_resa = random.randint(5, 10)  # Entre 5 et 10 réservations aléatoires
    chosen_events = df_events_passes.sample(n=min(num_resa, len(df_events_passes)), replace=False)  

    for _, event_row in chosen_events.iterrows():
        res_id = str(uuid.uuid4())  # UUID unique pour la réservation
        created_at = random_date(start_date, end_date).isoformat()

        reservation = {
            'id': res_id,
            'userId': user_row['accountId'],  
            'eventId': event_row['id'],  
            'createdAt': created_at,
            'updatedAt': created_at
        }
        reservations.append(reservation)


df_reservations = pd.DataFrame(reservations)
```
